```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata_df = pd.read_csv(mouse_metadata_path)
study_results_df = pd.read_csv(study_results_path)


# Combine the data into a single dataset
combined_data_df = pd.merge(mouse_metadata_df,study_results_df,
                                 how='outer', on='Mouse ID')
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
 
    
# Start by getting the last (greatest) timepoint for each mouse
Mice_Group2=Reduced_data_df.groupby(["Mouse ID"])
Max_timepoint=Mice_Group2[["Timepoint"]].max()
Max_timepoint
```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
with_max_timepoint_df=pd.merge(Max_timepoint,Reduced_data_df,how='left', on=['Mouse ID', 'Timepoint'])
with_max_timepoint_df.head()
```

```python
Max_timepoint
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list = []



for drug in treatment_list:
    Max_tumor_volume=with_max_timepoint_df.loc[with_max_timepoint_df["Drug Regimen"]==drug,"Tumor Volume (mm3)"]
    tumor_vol_list.append(Max_tumor_volume)
print(Max_tumor_volume)



Capomulin_tumor_volume=with_max_timepoint_df.loc[with_max_timepoint_df["Drug Regimen"]=="Capomulin","Tumor Volume (mm3)"]
Ramicane_tumor_volume=with_max_timepoint_df.loc[with_max_timepoint_df["Drug Regimen"]=="Ramicane","Tumor Volume (mm3)"]
Infubinol_tumor_volume=with_max_timepoint_df.loc[with_max_timepoint_df["Drug Regimen"]=="Infubinol","Tumor Volume (mm3)"]
Ceftamin_tumor_volume=with_max_timepoint_df.loc[with_max_timepoint_df["Drug Regimen"]=="Ceftamin","Tumor Volume (mm3)"]

    
```

```python

plt.boxplot(tumor_vol_list,labels=treatment_list)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
DF_Capomulin=Reduced_data_df[Reduced_data_df['Drug Regimen']=='Capomulin']
DF_Capomulin
Average=DF_Capomulin.groupby(["Mouse ID"]).mean()
Average["Tumor Volume (mm3)"]      

plt.scatter(Average['Weight (g)'], Average['Tumor Volume (mm3)'])

plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(Average['Weight (g)'], Average['Tumor Volume (mm3)'])
regress_values = Average['Weight (g)'] * slope + intercept

print(regress_values)

line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
print(line_eq)
plt.plot([15.06, 25.11],[35.86,45.41],"r-")

# Print r value
print(f"The The correlation between mouse weight and the average tumor volume is: {rvalue}")
```
